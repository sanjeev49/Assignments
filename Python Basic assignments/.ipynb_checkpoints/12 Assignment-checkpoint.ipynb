## Answer 1

* PdfFileReader() should be open in binary mode by giving the argument 'rb' while PdfFileWriter() should be opned in write-binary mode by giving the argument `wb`.


```python
import PyPDF2
file = open(r'C:\Users\Sanjeev\Desktop\automate2e_SampleCh7.pdf', 'rb')
pdfreader  = PyPDF2.PdfFileReader(file)
pdfreader.numPages
```

```python
pageobj = pdfreader.getPage(0)
pageobj.extractText()
```

## Answer 2

```python
# To get a page from page 5 from pdf
pageobj = pdfreader.getPage(6)
pageobj.extractText()
```
